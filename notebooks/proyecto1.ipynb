```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargamos el dataset
df = pd.read_csv('wfh_burnout_dataset.csv')

# Definimos X e y
X = df.drop(['user_id', 'burnout_risk', 'burnout_score'], axis=1)
y = df['burnout_risk']

print("Datos cargados correctamente.")
print(f"Forma de X (Entradas): {X.shape}")
print(f"Ejemplos de etiquetas (y): {y.unique()}")

X.head()
```

```python
# Cantidad de datos y tipos
print(df.info())

# Estadísticas principales
print(df.describe())

# Dtos nulos
print(f"Datos faltantes: {df.isnull().sum().sum()}")
```

```python
# 1. Distribución del riesgo de Burnout
plt.figure(figsize=(8,5))
sns.countplot(x='burnout_risk', data=df, palette='viridis')
plt.title('Distribución del Riesgo de Burnout (Variable y)')
plt.show()

# 2. Correlación entre variables numéricas
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlación')
plt.show()
```
